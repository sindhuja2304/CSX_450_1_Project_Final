```r
library(pryr)
library(tidyverse)
library(repr)
library(ggplot2)
library(corrplot)
```

```r
AbaloneURL<- "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
Abalone <- read.csv(url(AbaloneURL), sep = ",", header = FALSE)
```

```r
Abalone <- Abalone %>% rename(Sex = V1, Length = V2, Diameter = V3, Height = V4, Whole = V5, Shucked = V6, Viscera = V7, Shell = V8, Rings = V9)
```
